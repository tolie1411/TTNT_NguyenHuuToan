```python
def tron_hai_mang(a, b):
    ket_qua = []
    n1, n2 = len(a), len(b)
    min_len = min(n1, n2)


    for i in range(min_len):
        ket_qua.append(a[i] + b[i])


    if n1 > n2:
        ket_qua.extend(a[min_len:])
    else:
        ket_qua.extend(b[min_len:])

    return ket_qua


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print("Kết quả trộn:", tron_hai_mang(a, b))
```

```python
import random

def bai_14():
    m, n = 3, 4


    a = [[random.randint(1, 100) for _ in range(n)] for _ in range(m)]
    print("Ma trận a:")
    for row in a: print(row)


    k = 1
    print(f"Dòng {k}:", a[k])


    print(f"Cột {k}:", [a[i][k] for i in range(m)])


    max_sum = -1
    dong_max = 0
    for i, row in enumerate(a):
        if sum(row) > max_sum:
            max_sum = sum(row)
            dong_max = i
    print(f"Dòng có tổng lớn nhất là dòng {dong_max} với tổng {max_sum}")


    min_tich = float('inf')
    cot_min = 0
    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]
        if tich < min_tich:
            min_tich = tich
            cot_min = j
    print(f"Cột có tích nhỏ nhất là cột {cot_min}")


    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(0, m, 2):
        for j in range(1, n, 2):
            print(f"a[{i}][{j}] = {a[i][j]}", end=" ")
    print()


    ds_chan_dong_le = [a[i][j] for i in range(1, m, 2) for j in range(n) if a[i][j] % 2 == 0]
    tbc_g = sum(ds_chan_dong_le) / len(ds_chan_dong_le) if ds_chan_dong_le else 0
    print("TBC phần tử chẵn thuộc dòng lẻ:", tbc_g)


    bien = []
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                bien.append(a[i][j])
    print("TBC phần tử biên:", sum(bien)/len(bien))


    khong_bien = [a[i][j] for i in range(1, m-1) for j in range(1, n-1)]
    if khong_bien:
        print("TBC phần tử không thuộc biên:", sum(khong_bien)/len(khong_bien))
    else:
        print("Không có phần tử nằm trong biên (ma trận quá nhỏ)")

bai_14()
```

```python
from datetime import datetime

class SinhVien:
    def __init__(self, ma, ten, nam_sinh, diem_tb):
        self.ma = ma[:10]
        self.ten = ten[:20]
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

def quan_ly_sinh_vien():

    danh_sach = [
        SinhVien("01DH0001", "Phan Anh Lan", 2004, 7.5),
        SinhVien("02DH0005", "Nguyen Van A", 2006, 4.0),
        SinhVien("03CD0001", "Tran Thi Lan", 2003, 8.0),
        SinhVien("04DH1234", "Phan Thanh B", 2006, 6.0)
    ]


    len_lop = [sv for sv in danh_sach if sv.diem_tb >= 5]
    print(f"Số SV đủ điều kiện lên lớp: {len(len_lop)}")


    nam_hien_tai = 2026
    print("SV đủ 20 tuổi:")
    for sv in danh_sach:
        if nam_hien_tai - sv.nam_sinh == 20:
            print(f"- {sv.ten} ({sv.nam_sinh})")


    sv_dh = [sv for sv in danh_sach if sv.ma[2:4] == "DH"]
    print(f"Số SV hệ Đại học: {len(sv_dh)}")


    count_lan = sum(1 for sv in danh_sach if sv.ten.split()[-1] == "Lan")
    print(f"Số SV tên Lan: {count_lan}")


    count_phan = sum(1 for sv in danh_sach if sv.ten.startswith("Phan"))
    print(f"Số SV họ Phan: {count_phan}")

quan_ly_sinh_vien()
```
