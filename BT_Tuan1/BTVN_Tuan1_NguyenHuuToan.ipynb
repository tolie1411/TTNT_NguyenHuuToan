```python
def tron_2_mang(a,b):
  min_len=len(a) if len(a)<len(b) else len(b)
  c=b if len(b)>len(a) else a
  for i in range(min_len):
    c[i]=b[i] if b[i]<a[i] else a[i]
  return c
a=[3,9,1,4]
b=[2,7,4,3,2,8]
print(tron_2_mang(a,b))
```

```python
import copy


def tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

def chuyen_am_thanh_tuyet_doi(a, m, n):
    for i in range(m):
        for j in range(n):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])

def thay_chan_bang_x(a, m, n, x):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x

def kiem_tra_toan_chan(a, m, n):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 != 0:
                return False
    return True

def kiem_tra_doi_xung(a, n):
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

def kiem_tra_cheo_chinh_tang_dan(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i + 1][i + 1]:
            return False
    return True

def xuat_tam_giac_duoi_cheo_phu(a, n):
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(f"{a[i][j]}\t", end="")
            else:
                print("\t", end="")
        print()

def kiem_tra_cheo_phu_giam_dan(a, n):
    for i in range(n - 1):
        if a[i][n - 1 - i] <= a[i + 1][n - 1 - (i + 1)]:
            return False
    return True


def in_ma_tran(a):
    for row in a:
        for val in row:
            print(f"{val}\t", end="")
        print()
    print("-" * 30)


if __name__ == "__main__":

    a = [
        [ 2, -4,  6,  8],
        [-3,  4, -7,  1],
        [ 9,  2,  6, -6],
        [-1,  0,  3,  7]
    ]
    m = 4
    n = 4

    print("=== MA TRẬN BAN ĐẦU ===")
    in_ma_tran(a)


    tong = tong_tam_giac_tren_cheo_phu(a, n)
    print(f"1. Tổng các phần tử thuộc tam giác trên chéo phụ: {tong}\n")


    kt_chan = kiem_tra_toan_chan(a, m, n)
    print(f"2. Ma trận có toàn số chẵn không? {'Có' if kt_chan else 'Không'}\n")


    kt_dx = kiem_tra_doi_xung(a, n)
    print(f"3. Ma trận có đối xứng không? {'Có' if kt_dx else 'Không'}\n")


    kt_tang = kiem_tra_cheo_chinh_tang_dan(a, n)
    print(f"4. Đường chéo chính có tăng dần không? {'Có' if kt_tang else 'Không'}\n")


    kt_giam = kiem_tra_cheo_phu_giam_dan(a, n)
    print(f"5. Đường chéo phụ có giảm dần không? {'Có' if kt_giam else 'Không'}\n")


    print("6. Các phần tử thuộc tam giác dưới chéo phụ:")
    xuat_tam_giac_duoi_cheo_phu(a, n)
    print()


    a_copy = copy.deepcopy(a)

    chuyen_am_thanh_tuyet_doi(a_copy, m, n)
    print("7. Ma trận sau khi chuyển các số âm thành trị tuyệt đối:")
    in_ma_tran(a_copy)


    thay_chan_bang_x(a_copy, m, n, 99)
    print("8. Ma trận sau khi thay các số chẵn bằng 99:")
    in_ma_tran(a_copy)
```
