```python
#bài 3&4
import math

class TicTacToeNxN:
    def __init__(self, n=5, k=4):
        self.N = n
        self.K = k
        self.board = [['_' for _ in range(self.N)] for _ in range(self.N)]
        self.player = 'X'
        self.ai = 'O'

    def print_board(self):
        print("\n  " + " ".join([str(i) for i in range(self.N)]))
        for i in range(self.N):
            print(f"{i} " + " ".join(self.board[i]))
        print()

    def get_available_moves(self):
        return [(i, j) for i in range(self.N) for j in range(self.N) if self.board[i][j] == '_']

    def check_win(self, board, piece):
        for i in range(self.N):
            for j in range(self.N):
                if board[i][j] == piece:
                    if j + self.K <= self.N and all(board[i][j+d] == piece for d in range(self.K)): return True
                    if i + self.K <= self.N and all(board[i+d][j] == piece for d in range(self.K)): return True
                    if i + self.K <= self.N and j + self.K <= self.N and all(board[i+d][j+d] == piece for d in range(self.K)): return True
                    if i + self.K <= self.N and j - self.K >= -1 and all(board[i+d][j-d] == piece for d in range(self.K)): return True
        return False

    def is_terminal(self, board):
        if self.check_win(board, self.ai): return 10
        if self.check_win(board, self.player): return -10
        if not self.get_available_moves(): return 0
        return None

    def evaluate(self, board):
        if self.check_win(board, self.ai): return 1000
        if self.check_win(board, self.player): return -1000
        return 0

    # Dùng Alpha-Beta cho tốc độ nhanh (Bài 4)
    def alpha_beta(self, board, depth, alpha, beta, is_maximizing, max_depth=4):
        terminal_val = self.is_terminal(board)
        if terminal_val is not None: return terminal_val
        if depth == max_depth: return self.evaluate(board)

        if is_maximizing:
            max_eval = -math.inf
            for move in self.get_available_moves():
                board[move[0]][move[1]] = self.ai
                eval_val = self.alpha_beta(board, depth + 1, alpha, beta, False, max_depth)
                board[move[0]][move[1]] = '_'
                max_eval = max(max_eval, eval_val)
                alpha = max(alpha, eval_val)
                if beta <= alpha: break
            return max_eval
        else:
            min_eval = math.inf
            for move in self.get_available_moves():
                board[move[0]][move[1]] = self.player
                eval_val = self.alpha_beta(board, depth + 1, alpha, beta, True, max_depth)
                board[move[0]][move[1]] = '_'
                min_eval = min(min_eval, eval_val)
                beta = min(beta, eval_val)
                if beta <= alpha: break
            return min_eval

    def find_best_move(self, search_depth=4):
        best_val = -math.inf
        best_move = (-1, -1)
        for move in self.get_available_moves():
            self.board[move[0]][move[1]] = self.ai
            move_val = self.alpha_beta(self.board, 0, -math.inf, math.inf, False, max_depth=search_depth)
            self.board[move[0]][move[1]] = '_'
            if move_val > best_val:
                best_val = move_val
                best_move = move
        return best_move

# === PHẦN CHẠY THỬ (MAIN LOOP) ===
def play_game():
    # Khởi tạo bàn cờ 5x5, cần 4 quân liên tiếp để thắng
    game = TicTacToeNxN(n=5, k=4)
    print("=== BẮT ĐẦU TRÒ CHƠI TICTACTOE NxN ===")
    print(f"Bàn cờ {game.N}x{game.N}. Cần {game.K} quân liên tiếp để thắng.")
    game.print_board()

    while True:
        # Lượt người chơi (X)
        try:
            user_input = input("Nhập nước đi của bạn (hàng cột, cách nhau bởi khoảng trắng - VD: 1 2): ")
            row, col = map(int, user_input.split())
            if (row, col) not in game.get_available_moves():
                print("Nước đi không hợp lệ hoặc ô đã được đánh! Vui lòng nhập lại.")
                continue
        except (ValueError, IndexError):
            print("Cú pháp sai. Hãy nhập hai số nguyên hợp lệ.")
            continue

        game.board[row][col] = game.player
        game.print_board()

        terminal = game.is_terminal(game.board)
        if terminal is not None:
            break

        # Lượt AI (O)
        print("AI đang suy nghĩ...")
        # Ở bàn cờ to, search_depth không nên để quá sâu (3 hoặc 4 là vừa) để tránh đợi lâu
        best_move = game.find_best_move(search_depth=3)
        if best_move != (-1, -1):
            game.board[best_move[0]][best_move[1]] = game.ai
            print(f"AI đánh vào ô: {best_move[0]} {best_move[1]}")
            game.print_board()

        terminal = game.is_terminal(game.board)
        if terminal is not None:
            break

    # Kết thúc
    if terminal == -10:
        print("🎉 Chúc mừng bạn đã thắng!")
    elif terminal == 10:
        print("💻 AI đã thắng!")
    else:
        print("🤝 Hòa!")

# Bấm chạy lệnh này
play_game()
```

```python
#bài 5
import chess
import math


def evaluate_board(board):
    if board.is_checkmate():
        return -9999 if board.turn else 9999
    if board.is_stalemate() or board.is_insufficient_material():
        return 0

    piece_values = {
        chess.PAWN: 1, chess.KNIGHT: 3, chess.BISHOP: 3,
        chess.ROOK: 5, chess.QUEEN: 9, chess.KING: 0
    }

    evaluation = 0
    for square in chess.SQUARES:
        piece = board.piece_at(square)
        if piece:
            value = piece_values[piece.piece_type]
            if piece.color == chess.WHITE:
                evaluation += value
            else:
                evaluation -= value
    return evaluation


def alpha_beta_chess(board, depth, alpha, beta, is_maximizing):
    if depth == 0 or board.is_game_over():
        return evaluate_board(board)

    if is_maximizing:
        max_eval = -math.inf
        for move in board.legal_moves:
            board.push(move)
            eval_score = alpha_beta_chess(board, depth - 1, alpha, beta, False)
            board.pop()

            max_eval = max(max_eval, eval_score)
            alpha = max(alpha, eval_score)
            if beta <= alpha: break
        return max_eval
    else:
        min_eval = math.inf
        for move in board.legal_moves:
            board.push(move)
            eval_score = alpha_beta_chess(board, depth - 1, alpha, beta, True)
            board.pop()

            min_eval = min(min_eval, eval_score)
            beta = min(beta, eval_score)
            if beta <= alpha: break
        return min_eval

def get_ai_move(board, depth=3):
    best_move = None
    best_value = -math.inf

    for move in board.legal_moves:
        board.push(move)
        board_value = alpha_beta_chess(board, depth - 1, -math.inf, math.inf, False)
        board.pop()

        if board_value > best_value:
            best_value = board_value
            best_move = move

    return best_move

print("=== DEMO AI CỜ VUA BẰNG ALPHA-BETA ===")
board = chess.Board()


for _ in range(5):
    print("\nBàn cờ hiện tại:")
    print(board)
    print("----------------")

    print("AI (Trắng) đang tính toán...")
    ai_move = get_ai_move(board, depth=3)
    print(f"=> AI quyết định đi: {ai_move}")
    board.push(ai_move)



    black_move = list(board.legal_moves)[0]
    print(f"=> Đen đi: {black_move}")
    board.push(black_move)

print("\nBàn cờ sau 5 lượt:")
print(board)
```

```python
#bài 6
import ipywidgets as widgets
from IPython.display import display, clear_output
import math

class TicTacToeColab:
    def __init__(self):
        self.board = [['_' for _ in range(3)] for _ in range(3)]
        self.player = 'X'
        self.ai = 'O'
        self.buttons = [[None for _ in range(3)] for _ in range(3)]
        self.output = widgets.Output()
        self.create_ui()

    def create_ui(self):

        grid_rows = []
        for i in range(3):
            row_buttons = []
            for j in range(3):

                btn = widgets.Button(description=' ', layout=widgets.Layout(width='60px', height='60px'))
                btn.style.button_color = '#e0e0e0'
                btn.style.font_weight = 'bold'


                btn.on_click(lambda b, r=i, c=j: self.player_move(r, c))

                self.buttons[i][j] = btn
                row_buttons.append(btn)

            grid_rows.append(widgets.HBox(row_buttons))


        self.reset_btn = widgets.Button(description='Chơi Lại', button_style='info')
        self.reset_btn.on_click(self.reset_game)


        ui = widgets.VBox(grid_rows + [self.reset_btn, self.output])
        display(ui)

    def player_move(self, row, col):

        if self.board[row][col] == '_' and not self.is_game_over():
            self.board[row][col] = self.player
            self.buttons[row][col].description = self.player
            self.buttons[row][col].style.button_color = 'lightblue'

            if not self.check_winner_and_draw():
                with self.output:
                    clear_output(wait=True)
                    print("AI đang tính toán...")
                self.ai_move()

    def ai_move(self):
        best_score = -math.inf
        best_move = None


        for i in range(3):
            for j in range(3):
                if self.board[i][j] == '_':
                    self.board[i][j] = self.ai
                    score = self.alpha_beta(self.board, 0, -math.inf, math.inf, False)
                    self.board[i][j] = '_'
                    if score > best_score:
                        best_score = score
                        best_move = (i, j)

        if best_move:
            self.board[best_move[0]][best_move[1]] = self.ai
            self.buttons[best_move[0]][best_move[1]].description = self.ai
            self.buttons[best_move[0]][best_move[1]].style.button_color = '#ffb3b3'

            with self.output:
                clear_output(wait=True)
            self.check_winner_and_draw()


    def alpha_beta(self, board, depth, alpha, beta, is_maximizing):
        result = self.check_win_state(board)
        if result == self.ai: return 10 - depth
        if result == self.player: return -10 + depth
        if result == 'Tie': return 0

        if is_maximizing:
            max_eval = -math.inf
            for i in range(3):
                for j in range(3):
                    if board[i][j] == '_':
                        board[i][j] = self.ai
                        eval_score = self.alpha_beta(board, depth + 1, alpha, beta, False)
                        board[i][j] = '_'
                        max_eval = max(max_eval, eval_score)
                        alpha = max(alpha, eval_score)
                        if beta <= alpha: break
            return max_eval
        else:
            min_eval = math.inf
            for i in range(3):
                for j in range(3):
                    if board[i][j] == '_':
                        board[i][j] = self.player
                        eval_score = self.alpha_beta(board, depth + 1, alpha, beta, True)
                        board[i][j] = '_'
                        min_eval = min(min_eval, eval_score)
                        beta = min(beta, eval_score)
                        if beta <= alpha: break
            return min_eval

    def check_win_state(self, board):

        for i in range(3):
            if board[i][0] == board[i][1] == board[i][2] != '_': return board[i][0]
            if board[0][i] == board[1][i] == board[2][i] != '_': return board[0][i]

        if board[0][0] == board[1][1] == board[2][2] != '_': return board[0][0]
        if board[0][2] == board[1][1] == board[2][0] != '_': return board[0][2]


        if all(board[i][j] != '_' for i in range(3) for j in range(3)): return 'Tie'
        return None

    def is_game_over(self):
        return self.check_win_state(self.board) is not None

    def check_winner_and_draw(self):
        res = self.check_win_state(self.board)
        if res:
            with self.output:
                clear_output(wait=True)
                if res == self.player:
                    print("🎉 Kết thúc: Bạn (X) đã thắng!")
                elif res == self.ai:
                    print("💻 Kết thúc: AI (O) đã thắng!")
                elif res == 'Tie':
                    print("🤝 Kết thúc: Trò chơi hòa!")
            return True
        return False

    def reset_game(self, b=None):
        self.board = [['_' for _ in range(3)] for _ in range(3)]
        for i in range(3):
            for j in range(3):
                self.buttons[i][j].description = ' '
                self.buttons[i][j].style.button_color = '#e0e0e0'
        with self.output:
            clear_output()


game = TicTacToeColab()
```
