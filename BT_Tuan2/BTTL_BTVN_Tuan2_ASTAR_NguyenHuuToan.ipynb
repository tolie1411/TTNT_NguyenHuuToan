```python
#bài 2(A* bài tập về nhà)
import heapq

class GraphTSP:
    def __init__(self, num_nodes):
        self.num_nodes = num_nodes

        self.adj_matrix = [[0] * num_nodes for _ in range(num_nodes)]

    def add_edge(self, u, v, weight):
        self.adj_matrix[u][v] = weight
        self.adj_matrix[v][u] = weight

def tsp_a_star(graph, start_node):
    num_nodes = graph.num_nodes


    initial_state = (0, 0, start_node, [start_node], (1 << start_node))

    pq = []
    heapq.heappush(pq, initial_state)

    best_cost = float('inf')
    best_path = []

    while pq:
        f_cost, g_cost, u, path, visited = heapq.heappop(pq)


        if visited == (1 << num_nodes) - 1:

            return_cost = graph.adj_matrix[u][start_node]
            if return_cost > 0:
                total_cost = g_cost + return_cost
                if total_cost < best_cost:
                    best_cost = total_cost
                    best_path = path + [start_node]
            continue


        for v in range(num_nodes):

            if not (visited & (1 << v)) and graph.adj_matrix[u][v] > 0:
                new_visited = visited | (1 << v)
                new_g_cost = g_cost + graph.adj_matrix[u][v]
                new_path = path + [v]


                h_cost = 0
                new_f_cost = new_g_cost + h_cost

                heapq.heappush(pq, (new_f_cost, new_g_cost, v, new_path, new_visited))

    return best_path, best_cost


g = GraphTSP(4)
g.add_edge(0, 1, 10)
g.add_edge(0, 2, 15)
g.add_edge(0, 3, 20)
g.add_edge(1, 2, 35)
g.add_edge(1, 3, 25)
g.add_edge(2, 3, 30)

start_city = 0
path, cost = tsp_a_star(g, start_city)

print("--- Kết quả Bài toán Người giao hàng bằng A* ---")
print(f"Đường đi tối ưu: {path}")
print(f"Tổng chi phí: {cost}")
```

```python
#Bài 2: A*(Bài ở lớp)
import heapq

class GraphAStar:
    def __init__(self):

        self.adj_list = {}

        self.heuristics = {}

    def add_edge(self, source, target, weight):

        if source not in self.adj_list:
            self.adj_list[source] = []
        self.adj_list[source].append((target, weight))

    def set_heuristic(self, node, h_value):

        self.heuristics[node] = h_value

    def find_path(self, start, goal):



        open_set = []
        heapq.heappush(open_set, (0, start))


        came_from = {}


        g_score = {start: 0}


        f_score = {start: self.heuristics.get(start, 0)}

        while open_set:

            current_f, current = heapq.heappop(open_set)


            if current == goal:
                return self.reconstruct_path(came_from, current)


            if current in self.adj_list:
                for neighbor, weight in self.adj_list[current]:

                    tentative_g_score = g_score[current] + weight


                    if tentative_g_score < g_score.get(neighbor, float('inf')):

                        came_from[neighbor] = current
                        g_score[neighbor] = tentative_g_score
                        f_score[neighbor] = tentative_g_score + self.heuristics.get(neighbor, 0)


                        heapq.heappush(open_set, (f_score[neighbor], neighbor))


        return []

    def reconstruct_path(self, came_from, current):

        path = [current]
        while current in came_from:
            current = came_from[current]
            path.append(current)
        path.reverse()
        return path
if __name__ == "__main__":
    graph = GraphAStar()


    graph.add_edge("A", "B", 7)
    graph.add_edge("A", "C", 15)
    graph.add_edge("B", "C", 23)
    graph.add_edge("B", "D", 54)
    graph.add_edge("C", "D", 12)
    graph.add_edge("D", "E", 39)


    graph.set_heuristic("A", 75)
    graph.set_heuristic("B", 36)
    graph.set_heuristic("C", 32)
    graph.set_heuristic("D", 11)
    graph.set_heuristic("E", 0)

    start_node = "A"
    goal_node = "E"

    path = graph.find_path(start_node, goal_node)

    if path:
        print(f"Đường đi ngắn nhất từ {start_node} đến {goal_node} là: {' -> '.join(path)}")
    else:
        print("Không tìm thấy đường đi!")
```
